```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("/content/PUBG game dataset.csv")
data.head()
```

```python
data.shape
```

```python
data.columns
data["Players joined"] = data.groupby("MatchId")["MatchId"].transform("count")
```

```python
data.drop_duplicates(subset="ID")
```

```python
del data["Walk distance"],data["Swim distance"],data["Ride distance"],data["numGroups"]
```

```python
data.head(1)
```

```python
print(data["Headshot kills"].value_counts())
print(data["Kills"].isna().value_counts())
```

```python
data["Kills"] = pd.to_numeric(data["Kills"],errors = "coerce")
data["Headshot kills"] = pd.to_numeric(data["Headshot kills"],errors = "coerce")
```

```python
data['Headshot rate'] = np.where(data['Kills'] > 0, data['Headshot kills'] / data['Kills'], 0)
```

```python
data.drop(data[(data["Kills"] > 5)& (data["Headshot kills"]==1)].index, inplace = True)
```

```python
data.shape
```

```python
data["Killswithoutmoving"] = (data["Kills"]>0) & (data["Total distance"])
data.drop(data[data["Killswithoutmoving"]==True].index, inplace = True)
```

```python
del data["Killswithoutmoving"]
```

```python
data.shape
```

```python
data["Impossibleroadkills"] = data["Road kills"]>6
data.drop(data[data["Impossibleroadkills"]==True].index, inplace = True)
```

```python
del data["Impossibleroadkills"]
```

```python
data.drop(data[data["Kills"]>20].index,inplace = True)
```

```python
data.head()
```

```python
data.drop(data[data["Longest kill"]>500].index,inplace = True)
```

```python
data.drop(data[data["Weapons acquired"]>12].index,inplace=True)
```

```python
data.info()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data["Healings"] = data["Heals"] + data["Boosts"]
```

```python
data = data.drop(columns = ["ID","GroupID","MatchId","Heals","Boosts"])
```

```python
data
```

```python
x = data.drop(columns = "Win percentage")
y = data["Win percentage"]
```

```python
x = pd.get_dummies(x, columns=["Match type"])
```

```python
data["Match type"].value_counts()
```

```python
x
```

```python
y
```

```python
data["Win percentage"] = data["Win percentage"]*100
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.4, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
```

```python
data.head()
```

```python
model.fit(x_train,y_train)
```

```python
predictions = model.predict(x)
```

```python
data["Predicted win percentage"] = predictions*100
```

```python
data
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
RMSE = np.sqrt(mean_squared_error(y, predictions))
R2_score = r2_score(y, predictions)
print("Testing performance")
print(f"RMSE: {RMSE:.2f}")
print(f"R2_score: {R2_score:.2f}")
```
